```python
import pandas as pd
```

# import data

```python
area_spm = pd.read_csv('C:/Users/l_v_v/Documents/GitHub/time_series_curuai/datasets/Parameters Time series/area_SPM_time_series_cota.csv').drop(columns=['Unnamed: 0'])
precipitation = pd.read_csv('C:/Users/l_v_v/Documents/GitHub/time_series_curuai/datasets/Parameters Time series/mean_precip_acum_cota.csv')
discharge = pd.read_csv('C:/Users/l_v_v/Documents/GitHub/time_series_curuai/datasets/Parameters Time series/mean_watershed_discharge_cota.csv').drop(columns=['Unnamed: 0'])
wind = pd.read_csv('C:/Users/l_v_v/Documents/GitHub/time_series_curuai/datasets/Parameters Time series/mean_watershed_wind_cota.csv')
```

# organize dataframes

```python
print('colunms area', area_spm.columns)
print('colunms precipitation', precipitation.columns)
print('colunms discharge', discharge.columns)
print('colunms wind', wind.columns)
```

```python
area_spm = area_spm[['area_km2', 'mean_SPM', 'max_SPM', 'min_SPM', 'stdDev_SPM',
       'time_start', 'year', 'WATER_PERIOD']].copy()
precipitation = precipitation[['mean_precipitation', 
       'time_start', 'year']].copy()
discharge = discharge[['media_vazao',  'year',
       'time_start']].copy()
wind = wind[['mean_u__wind', 'mean_v__wind', 
       'time_finish', 'time_start', 'year']].copy()
```

```python
area_spm.columns = ['area_km2', 'mean_SPM', 'max_SPM', 'min_SPM', 'stdDev_SPM', 
        'time_start', 'year', 'water_period']
precipitation.columns = ['mean_precipitation', 'time_start', 'year']
discharge.columns = [ 'mean_discharge',  'year',
       'time_start']
wind.columns = ['mean_u__wind', 'mean_v__wind', 
       'time_finish', 'time_start', 'year']
```

# merge dataframes

```python
precipitation
```

```python
df_pt1 =  pd.merge(area_spm, precipitation, how="left", on=["time_start", "year"])
df_pt1
```

```python
df_pt2 =  pd.merge(discharge, wind, how="outer", on=["time_start", "year"])
df_pt2
```

```python
df_total = pd.merge(df_pt1, df_pt2, how="left", on=["time_start", "year"])
df_total
```

```python
df_total.to_csv('C:/Users/l_v_v/Documents/GitHub/time_series_curuai/datasets/Parameters Time series/merged_df/df_merged_cota.csv')
```
