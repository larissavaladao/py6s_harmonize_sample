```python
import pandas as pd
```

# import data

```python
area_spm = pd.read_csv('C:/Users/l_v_v/Documents/GitHub/py6s_harmonize_sample/datasets/Parameters Time series/area_SPM_time_series.csv').drop(columns=['Unnamed: 0'])
precipitation = pd.read_csv('C:/Users/l_v_v/Documents/GitHub/py6s_harmonize_sample/datasets/Parameters Time series/mean_precip_acum.csv')
discharge = pd.read_csv('C:/Users/l_v_v/Documents/GitHub/py6s_harmonize_sample/datasets/Parameters Time series/mean_watershed_discharge.csv')
wind = pd.read_csv('C:/Users/l_v_v/Documents/GitHub/py6s_harmonize_sample/datasets/Parameters Time series/mean_watershed_wind.csv')
```

# organize dataframes

```python
print('colunms area', area_spm.columns)
print('colunms precipitation', precipitation.columns)
print('colunms discharge', discharge.columns)
print('colunms wind', wind.columns)
```

```python
area_spm = area_spm[['area_km2', 'mean_SPM', 'max_SPM', 'min_SPM', 'stdDev_SPM', 
       'month_init', 'year', 'WATER_PERIOD']].copy()
precipitation = precipitation[['mean_precipitation', 'month', 'year']].copy()
discharge = discharge[['water_period', 'year', 'start_month', 'Maxima', 'Minima', 'Media']].copy()
wind = wind[['mean_u__wind', 'mean_v__wind', 'month', 'year']].copy()
```

```python
area_spm.columns = ['area_km2', 'mean_SPM', 'max_SPM', 'min_SPM', 'stdDev_SPM', 
       'month_init', 'year', 'water_period']
precipitation.columns = ['mean_precipitation', 'month_init', 'year']
discharge.columns = ['water_period', 'year', 'month_init', 'discharge_max', 'discharge_min', 'discharge_mean']
wind.columns = ['mean_u__wind', 'mean_v__wind', 'month_init', 'year']
```

```python
def add_period(row):
    if row['month_init'] == 1 or row['month_init']== 2 or row['month_init'] == 3:
        val = 'R'
    elif row['month_init'] == 4 or row['month_init'] == 5 or row['month_init'] == 6:
        val = 'HW'
    elif row['month_init'] == 7 or row['month_init'] == 8 or row['month_init'] == 9:
        val = 'F'
    else:
        val = 'LW'
    return val

precipitation['water_period'] = precipitation.apply(add_period, axis=1)
wind['water_period'] = wind.apply(add_period, axis=1)
```

# merge dataframes

```python
df_pt1 =  pd.merge(area_spm, precipitation, how="inner", on=["month_init", "year",'water_period'])
df_pt1
```

```python
df_pt2 =  pd.merge(discharge, wind, how="inner", on=["month_init", "year",'water_period'])
df_pt2
```

```python
df_total = pd.merge(df_pt1, df_pt2, how="inner", on=["month_init", "year",'water_period'])
df_total
```

```python
df_total.to_csv('C:/Users/l_v_v/Documents/GitHub/py6s_harmonize_sample/datasets/Parameters Time series/merged_df/df_merged.csv')
```
